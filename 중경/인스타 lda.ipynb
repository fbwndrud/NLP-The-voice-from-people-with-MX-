```python
import pandas as pd
```

```python
df = pd.read_csv('insta1.csv', index_col = 0)
```

```python
df.columns = ['date','time','locate','힐링','len','tag','url']
```

```python
df.locate.value_counts()
```

```python
df.date.value_counts()
```

```python
df = df.loc[df.tag != '[]']
```

```python
df = df.reset_index(drop=True)
```

```python
for idx in range(len(df.tag)):
    df['tag'][idx]=df.tag[idx][1:-1].replace("'", '').replace("'", '').split(',')
    if idx %100 == 0:
        print(idx/len(df.tag)*100)
```

```python
len(df.url.value_counts())
```

```python
df.loc[df.url== 'https://www.instagram.com/p/B1Os5wcBAJj/']
```

```python
#df.tag에 list로 된 애들이 들어가 있어야 
from gensim import corpora
dictionary = corpora.Dictionary(df.tag)
corpus = [dictionary.doc2bow(text) for text in df.tag]
# print(corpus[1]) # 수행된 결과에서 두번째 뉴스 출력. 첫번째 문서의 인덱스는 0
import gensim
NUM_TOPICS = 5 #20개의 토픽, k=20
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=4)
for topic in topics:
    print(topic)
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```
