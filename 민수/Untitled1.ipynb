```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns


%matplotlib inline
plt.figure(figsize=(15,15))
sns.heatmap(data = sd.corr(), annot=True, 
fmt = '.2f', linewidths=.5, cmap='Reds')
```
