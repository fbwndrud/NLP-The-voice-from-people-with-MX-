```python
import pandas as pd
df = pd.read_csv('news.csv', index_col=0)
# df.content = df.content.str.replace('\n','')
# df.content = df.content.str.replace('\xa0','')
```

```python
from konlpy.tag import Okt
```

```python
df = df.drop(columns = 'date.1')
```

```python
df['token'] = None
```

```python
okt = Okt()
taglist=df.copy()
```

```python
okt.pos?
```

```python
for idx, val in enumerate(df.content):
    try:
        taglist.iloc[idx][4] = okt.pos(val, join=True,stem=True)
        if idx %500 == 0:
            print(idx/len(df)*100)
    except:
        taglist.append(float('nan'))
```

```python
def tokenfilter(text) :
    tokens = []
    filter_tag = ['Noun','Adjective', 'Adverb']
    auxwords = ['못하', '아니', '않', '지만']
    for tag in text:
        if tag.split('/')[1] in filter_tag:
            tokens.append(tag)
        elif tag.split('/')[0] in auxwords:
            tokens.append(tag)
    return tokens
```

```python
for idx, val in enumerate(taglist.token):
    df.iloc[idx][4] = tokenfilter(val)
```

```python
df.to_csv('news_v1.csv')
```
